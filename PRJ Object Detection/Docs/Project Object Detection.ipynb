                                FACE DETECTION WITH CASCADE CLASSIFIER

```python
#Calls Cascade classifier for face detction
face_detector = cv2.CascadeClassifier('C:\\Users\\Dell\Dropbox\\My PC (DESKTOP-0O98POK)\\Documents\\Books & Notes\\M. Tech\\Acmegrade Files\\PRJ Object Detection\\Cascades\\haarcascade_frontalface_default.xml')
face_detector
```

```python

image = cv2.imread('C:\\Users\Dell\\Dropbox\\My PC (DESKTOP-0O98POK)\\Documents\\Books & Notes\M. Tech\\Acmegrade Files\\PRJ Object Detection\\Data Set\\GreatMinds.jpg')
image = cv2.resize(image, (800, 600)) # Resize image
image_gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
detections = face_detector.detectMultiScale(image_gray)
print (len(detections),'Faces detected ')
for (x, y, w, h) in detections:
    cv2.rectangle(image, (x, y), (x + w, y + h), (244,11,239), 3)
cv2.imshow('',image)
cv2.waitKey(0)
cv2.destroyAllWindows()
plt.show()

```

Parametric Study

```python
#Parameter: scaleFactor 
image = cv2.imread('C:\\Users\Dell\\Dropbox\\My PC (DESKTOP-0O98POK)\\Documents\\Books & Notes\M. Tech\\Acmegrade Files\\PRJ Object Detection\\Data Set\\GreatMinds.jpg')
image = cv2.resize(image, (800, 600)) # Resize image
image_gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
detections = face_detector.detectMultiScale(image_gray, scaleFactor = 1.09)
print (len(detections),'Faces detected ')
for (x, y, w, h) in detections:
    cv2.rectangle(image, (x, y), (x + w, y + h), (244,11,239), 2)
cv2.imshow('',image)
cv2.waitKey(0)
cv2.destroyAllWindows()
plt.show()


```

```python
#Parameter: minSize,maxSize
image = cv2.imread('C:\\Users\Dell\\Dropbox\\My PC (DESKTOP-0O98POK)\\Documents\\Books & Notes\M. Tech\\Acmegrade Files\\PRJ Object Detection\\Data Set\\GreatMinds.jpg')
image = cv2.resize(image, (800, 600)) # Resize image
image_gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
detections = face_detector.detectMultiScale(image_gray, scaleFactor = 1.09,
                                            minNeighbors=5,minSize=(30,30), maxSize=(50,50))
print (len(detections),'Faces detected ')
for (x, y, w, h) in detections:
    cv2.rectangle(image, (x, y), (x + w, y + h), (244,11,239), 2)
cv2.imshow('',image)
cv2.waitKey(0)
cv2.destroyAllWindows()
plt.show()

```

```python
#Eye Detector
eye_detector= cv2.CascadeClassifier('C:\\Users\\Dell\Dropbox\\My PC (DESKTOP-0O98POK)\\Documents\\Books & Notes\\M. Tech\\Acmegrade Files\\PRJ Object Detection\\Cascades\\haarcascade_eye.xml')
display (eye_detector)
```

```python
#Face + Eye Detector
image = cv2.imread('C:\\Users\Dell\\Dropbox\\My PC (DESKTOP-0O98POK)\\Documents\\Books & Notes\M. Tech\\Acmegrade Files\\PRJ Object Detection\\Data Set\\ColumbiaCrew.jpg')
display (image.shape)

image = cv2.resize(image, (2000,1600)) # Resize image
print(image.shape)
face_detections = face_detector.detectMultiScale(image, scaleFactor = 1.3, minSize = (30,30))
print (len(face_detections),'Faces detected ')
for (x, y, w, h) in face_detections:
  cv2.rectangle(image, (x, y), (x + w, y + h), (244,11,239), 2)

eye_detections = eye_detector.detectMultiScale(image, scaleFactor = 1.09, minNeighbors=5, minSize=(29,29), maxSize=(45,45))
print (len(eye_detections),'eyes detected ')
for (x, y, w, h) in eye_detections:
  #print(w, h)
  cv2.rectangle(image, (x, y), (x + w, y + h), (0,0,255), 2)
cv2.imshow('',image)
cv2.waitKey(0)
cv2.destroyAllWindows()
plt.show()



```
